## breast cancer

```python
import pandas as pd
import os

bc_path = "./data/wdbc.data"
bc_df = pd.read_csv(bc_path, header=None)

base_feats = [
    "radius", "texture", "perimeter", "area", "smoothness",
    "compactness", "concavity", "concave_points", "symmetry", "fractal_dimension"
]
feature_names = (
    [f"mean_{f}" for f in base_feats] +
    [f"se_{f}" for f in base_feats] +
    [f"worst_{f}" for f in base_feats]
)

X_bc = bc_df.iloc[:, 2:].copy()
X_bc.columns = feature_names

y_bc = bc_df.iloc[:, 1].map({"M": 1, "B": -1}).astype(int)
y_bc.name = "diagnosis"

os.makedirs("./data/breast-cancer", exist_ok=True)
X_bc.to_csv("./data/breast-cancer/X.csv", index=False)
y_bc.to_csv("./data/breast-cancer/y.csv", index=False)
```

## wine quality 

```python
import pandas as pd
import os

wine_path = "./data/wine_quality.csv"
wine_df = pd.read_csv(wine_path, sep=";")

X_wine = wine_df.iloc[:, :-1].copy()
X_wine.columns = [
    "fixed_acidity", "volatile_acidity", "citric_acid", "residual_sugar",
    "chlorides", "free_sulfur_dioxide", "total_sulfur_dioxide", "density",
    "pH", "sulphates", "alcohol"
]
y_wine = wine_df.iloc[:, -1].apply(lambda v: 1 if v > 5 else -1)
y_wine.name = "quality"

os.makedirs("./data/wine-quality", exist_ok=True)
X_wine.to_csv("./data/wine-quality/X.csv", index=False)
y_wine.to_csv("./data/wine-quality/y.csv", index=False)
```

## avila bible

```python
import pandas as pd
import os

avila_tr_path = "./data/avila-tr.txt"
avila_ts_path = "./data/avila-ts.txt"

feature_cols = [f"F{i}" for i in range(1, 11)]
all_cols = feature_cols + ["class"]

df_tr = pd.read_csv(avila_tr_path, header=None, names=all_cols)
df_ts = pd.read_csv(avila_ts_path, header=None, names=all_cols)

# ignore train and test sets and merge them
avila_df = pd.concat([df_tr, df_ts], ignore_index=True)

X_avila = avila_df[feature_cols].copy()

y_avila = avila_df["class"].apply(lambda c: 1 if c == "A" else -1).astype(int)
y_avila.name = "class"

os.makedirs("./data/avila", exist_ok=True)
X_avila.to_csv("./data/avila/X.csv", index=False)
y_avila.to_csv("./data/avila/y.csv", index=False)
```

## magic telescope

```python
import pandas as pd
import os

# local loading

magic_path = "./data/magic04.data"

feature_cols = [
    "fLength", "fWidth", "fSize", "fConc", "fConc1",
    "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist"
]
all_cols = feature_cols + ["class"]

magic_df = pd.read_csv(magic_path, header=None, names=all_cols)

X_magic = magic_df[feature_cols].copy()
y_magic = magic_df["class"].map({"g": 1, "h": -1}).astype(int)
y_magic.name = "class"

os.makedirs("./data/magic", exist_ok=True)
X_magic.to_csv("./data/magic/X.csv", index=False)
y_magic.to_csv("./data/magic/y.csv", index=False)
```

## swarm behavious

```python
flocking_path = "./data/Flocking.csv"
flocking_df = pd.read_csv(flocking_path)

flocking_df.columns = flocking_df.columns.str.strip()

target_col = flocking_df.columns[-1]

X_flocking = flocking_df.iloc[:, :-1].copy()
y_raw = flocking_df[target_col]

if set(pd.unique(y_raw)).issubset({0, 1}):
    y_flocking = y_raw.map({0: -1, 1: 1}).astype(int)

elif set(pd.unique(y_raw)).issubset({-1, 1}):
    y_flocking = y_raw.astype(int)
else:

    positive_class = y_raw.value_counts().idxmax()
    y_flocking = y_raw.apply(lambda v: 1 if v == positive_class else -1).astype(int)

y_flocking.name = "class"

os.makedirs("./data/flocking", exist_ok=True)
X_flocking.to_csv("./data/flocking/X.csv", index=False)
y_flocking.to_csv("./data/flocking/y.csv", index=False)
```
